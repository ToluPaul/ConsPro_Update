```python
# Creating a list so as to be able to identify each faculty in relation with the skill they are most interested in
# A dictionary for essential skill
skilldict = {'FX':0, 'webdev':1, 'mobiledev':2, 'techwrite':3, 'UI_UX':4, 'graphdes':5, 'afm':6, 'conwrite':7,\
             'copywrite':8, 'digmark':9, 'Frontend':10, 'Backend':11, 'datascience':12}
skillcol = {'FX':'Black', 'webdev':'Red', 'mobiledev':'Green', 'techwrite':'Blue', 'UI_UX':'Magenta', 'graphdes':'Cyan', 'afm':'Yellow', 'conwrite':'Black',\
             'copywrite':'Yellow', 'digmark':'Green', 'Frontend':'Magenta', 'Backend':'Blue', 'datascience':'Red'}
```

```python
# A list of all dataframes sorted by the various skills interested
si = []
si.append(FX)
si.append(webdev)
si.append(mobiledev)
si.append(techwrite)
si.append(UI_UX)
si.append(graphdes)
si.append(afm)
si.append(conwrite)
si.append(copywrite)
si.append(digmark)
si.append(Frontend)
si.append(Backend)
si.append(datascience)
```

```python
# Creation of a function to automatically accept input based on the various skills
def myplot(skillinterest):
  skilldict = {'FX':0, 'webdev':1, 'mobiledev':2, 'techwrite':3, 'UI_UX':4, 'graphdes':5, 'afm':6, 'conwrite':7,\
             'copywrite':8, 'digmark':9, 'Frontend':10, 'Backend':11, 'datascience':12}
  skillcol = {'FX':'Black', 'webdev':'Red', 'mobiledev':'Green', 'techwrite':'Blue', 'UI_UX':'Magenta', 'graphdes':'Cyan', 'afm':'Yellow', 'conwrite':'Black',\
             'copywrite':'Yellow', 'digmark':'Green', 'Frontend':'Magenta', 'Backend':'Blue', 'datascience':'Red'}
  for skills in skillinterest:
    plt.hist(si[skilldict[skills]].Faculty, bins=40, label=skills, color=skillcol[skills])
  plt.legend(loc='upper left', bbox_to_anchor=(1,1))
  plt.xticks(rotation=90)
  plt.show()
```
